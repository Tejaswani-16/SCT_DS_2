# Data Collection

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = "C:/Users/HP/OneDrive/Desktop/python/sample_dataset.csv"
df=pd.read_csv(data)
print(df.head())
```

# Data Cleaning

```python
print("Missing values bedore cleaning", df.isnull().sum())
for col in ["Age", "Income", "Spending_Score"]:
    df.fillna(df[col].median(), inplace=True)
df.drop_duplicates(inplace=True)
print("\nMissing values after cleaning:\n", df.isnull().sum())
print("Shape after cleaning:",df.shape)
```

# Data Anlaysis

```python
print("Dataset Info:")
print(df.info())
print("Summary Statistics:")
print(df.describe(include="all"))
df.columns = df.columns.str.strip()
print(df.columns.tolist())
```

# Relationships

```python
# Age vs Spending Score
plt.figure(figsize=(6,4))
sns.scatterplot(x="Age", y="Spending_Score", data=df, hue="Gender", alpha=0.7)
plt.title("Age vs Spending Score by Gender")
plt.show()
```

```python
# Income vs Spending Score
plt.figure(figsize=(6,4))
sns.scatterplot(x="Income", y="Spending_Score", data=df, hue="Education", alpha=0.7)
plt.title("Income vs Spending Score by Education")
plt.show()
```

```python
# Gender vs Spending Score
plt.figure(figsize=(6,4))
sns.barplot(x="Gender", y="Spending_Score", data=df, ci=None, palette="pastel")
plt.title("Average Spending Score by Gender")
plt.show()
```

# Trends & Patterns

```python
# Education vs Spending Score
plt.figure(figsize=(7,4))
sns.barplot(x="Education", y="Spending_Score", data=df, hue="Education", palette="Set2", legend=False, errorbar=None)
plt.title("Average Spending Score by Education")
plt.xticks(rotation=30)
plt.show()
```

```python
# Gender vs Spending score
plt.figure(figsize=(6,4))
sns.barplot(x="Gender", y="Spending_Score", data=df, hue="Gender", palette="pastel", legend=False, errorbar=None)
plt.title("Average Spending Score by Gender")
plt.show()
```
